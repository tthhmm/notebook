```python
filename = '/home/nfs/mjmu/haiming/data/visibility/' + 'visibilityDataNoLagsNewPreds.csv'
content = pd.read_csv(filename)
print(content)
```

```python
vis_data = content.values
print(vis_data.shape)
```

```python
#look at the first row of data
vis_data[0, :]
```

```python
# get the label
label = content['KORD_VIS_0']
print(label)
```

```python
# convert label to label value
vis_label = label.values
print(vis_label)
print(vis_label.shape)
```

```python
time_data = vis_data[:, 0]
vis_data = vis_data[:, 1:-1]
```

```python
vis_data.shape, vis_label.reshape(-1, 1).shape
```

```python
vis_label_1hour_p = vis_label[4:]
print (vis_label_1hour_p.shape)
print(vis_label_1hour_p[10:20])
vis_label_1hour_p_f = np.append(vis_label_1hour_p, [np.nan]*4)
print(vis_label_1hour_p_f.shape)
print(vis_label_1hour_p_f)
```

```python
# zip vis_data and label together for nan value fitering
data_t = np.concatenate((vis_data, vis_label_1hour_p_f.reshape(-1, 1)), axis = 1)
print(data_t.shape)
```

```python
print(data_t[-1, -1])
```

```python
#change the type from object to float for calculation
data_tt = data_t.astype(float)
data_tt
```

```python
b = np.isnan(data_tt).any(axis=1)
```

```python
data_final = data_tt[~b]
data_final.shape
```

```python
data_final
```

```python
time_data = time_data[~b]
```

```python
d = data_final[:, 0:-1]
l = data_final[:, -1]
d.shape, l.shape
```

```python
'''very rough to pre-process the visibility data, save it as reference. We will create different training dataset for 
different type of machine learning'''

pickle_file = '/home/nfs/mjmu/haiming/data/visibility/' +  'rough_visibility.pickle'

try:
    f = open(pickle_file, 'wb')
    save = {
        'time': time_data,
        'data': d,
        'label': l,
    }
    pickle.dump(save, f, pickle.HIGHEST_PROTOCOL)
    f.close()
except Exception as e:
  print('Unable to save data to', pickle_file, ':', e)
  raise
```
